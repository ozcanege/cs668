```python
!pip install sklearn
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
print(boston)
```

```python
df_x = pd.DataFrame(boston.data, columns=boston.feature_names)
df_y = pd.DataFrame(boston.target)
```

```python
del df_x['B']
df_x
```

```python
print(boston.DESCR)
```

```python
# Initializing the Linear Regression
reg_linearRegression = linear_model.LinearRegression()
```

```python
# Data = %67 training , %33 test
x_train, x_test, y_train , y_test = train_test_split(df_x, df_y, test_size = 0.33, random_state = 42)
```

```python
# Training the data
reg_linearRegression.fit(x_train,y_train)
```

```python
# Predict the values from x_test
y_pred_linearRegression = reg_linearRegression.predict(x_test)
```

```python
y_pred_linearRegression
```

```python
# Actual values for x_test
y_test
```

```python
# Model Performance
# Ne kadar az olursa o kadar iyi
from sklearn.metrics import mean_squared_error
mean_squared_value_linearRegression = mean_squared_error(y_test,y_pred_linearRegression)
print(mean_squared_value_linearRegression)
```

```python
# Logistic Regression
reg_logisticRegression = linear_model.LogisticRegression()
```

```python
from sklearn import preprocessing
from sklearn import utils

lab_enc = preprocessing.LabelEncoder()
training_scores_encoded = lab_enc.fit_transform(y_train)

reg_logisticRegression.fit(x_train,training_scores_encoded)

# Predict the values from x_test
y_pred_logisticRegression = reg_logisticRegression.predict(x_test)
print(y_pred_logisticRegression)

mean_squared_value_logisticRegression = mean_squared_error(y_test,y_pred_logisticRegression)
print(mean_squared_value_logisticRegression)
```

```python
from sklearn.tree import DecisionTreeRegressor
reg_decisionTreeRegressor = DecisionTreeRegressor(random_state=0)
reg_decisionTreeRegressor.fit(x_train,y_train)
y_pred_decisionTreeRegressor = reg_decisionTreeRegressor.predict(x_test)
y_pred_decisionTreeRegressor
```

```python
mean_squared_value_decisionTreeRegressor = mean_squared_error(y_test,y_pred_decisionTreeRegressor)
mean_squared_value_decisionTreeRegressor
```

```python

from sklearn.tree import DecisionTreeClassifier
reg_decisionTreeClassifier = DecisionTreeClassifier(random_state=0)
reg_decisionTreeClassifier.fit(x_train,training_scores_encoded)
y_pred_decisionTreeClassifier = reg_decisionTreeClassifier.predict(x_test)
y_pred_decisionTreeClassifier
```

```python
mean_squared_value_decisionTreeClassifier = mean_squared_error(y_test,y_pred_decisionTreeClassifier)
mean_squared_value_decisionTreeClassifier
```

```python
from sklearn.svm import SVC
from sklearn.svm import SVR
reg_svm = SVC(kernel='linear')
reg_svm.fit(x_train,training_scores_encoded)
y_pred_svm = reg_svm.predict(x_test)
y_pred_svm

mean_squared_value_svm = mean_squared_error(y_test,y_pred_svm)
mean_squared_value_svm
```

```python
from sklearn.naive_bayes import GaussianNB
reg_naiveBayes = GaussianNB()
reg_naiveBayes.fit(x_train,training_scores_encoded)
y_pred_naiveBayes = reg_naiveBayes.predict(x_test)
y_pred_naiveBayes

mean_squared_value_naiveBayes = mean_squared_error(y_test,y_pred_naiveBayes)
mean_squared_value_naiveBayes
```

```python
from sklearn.neighbors import KNeighborsClassifier
reg_knn = KNeighborsClassifier(n_neighbors=6) # Optimum value is 6 for this dataset
reg_knn.fit(x_train,training_scores_encoded)
y_pred_knn = reg_knn.predict(x_test)
y_pred_knn

mean_squared_value_knn = mean_squared_error(y_test,y_pred_knn)
mean_squared_value_knn
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor

reg_randomForestClassifier = RandomForestClassifier(max_depth=150, random_state=0)
reg_randomForestClassifier.fit(x_train,training_scores_encoded)
y_pred_randomForestClassifier = reg_randomForestClassifier.predict(x_test)
y_pred_randomForestClassifier

mean_squared_value_randomForestClassifier = mean_squared_error(y_test,y_pred_randomForestClassifier)
mean_squared_value_randomForestClassifier
```

```python
reg_randomForestRegressor = RandomForestRegressor()
reg_randomForestRegressor.fit(x_train,y_train)
y_pred_randomForestRegressor = reg_randomForestRegressor.predict(x_test)
y_pred_randomForestRegressor

mean_squared_value_randomForestRegressor = mean_squared_error(y_test,y_pred_randomForestRegressor)
mean_squared_value_randomForestRegressor
```
